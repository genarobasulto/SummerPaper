```julia
using CSV, DataFrames, Plots
```

```julia
#Import the data
product_data = CSV.read("KeepaExport-2023-05-29-ProductFinder.csv")
seller_data = CSV.read("KeepaExport-2023-05-29-TopSellerList.csv")
```

```julia
#Print the variables of the seller data
for n in names(seller_data)
    println("* ", n, "\n")
end
```

```julia
#Print the variables of the product data
for n in names(product_data)
    println("* ", n, "\n")
end
```

```julia
#Print one observation from the product data
for info in zip(names(product_data),product_data[1,:])
    println(info[1], " ", info[2])
end
```

```julia
#Print one observation from the seller data
for info in zip(names(seller_data),seller_data[1,:])
    println(info[1], " ", info[2])
end
```
